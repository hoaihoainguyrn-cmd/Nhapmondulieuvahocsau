```python
df['DT'].fillna(0, inplace=True)
df
```

```python
df['T1'].fillna(df['T1'].mean(), inplace=True)
df
```

```python
print(" SỐ LƯỢNG DỮ LIỆU THIẾU TRƯỚC KHI XỬ LÝ ")
cac_cot_diem = [
    'T1', 'L1', 'H1', 'S1', 'V1', 'X1', 'D1', 'N1',
    'T2', 'L2', 'H2', 'S2', 'V2', 'X2', 'D2', 'N2',
    'T6', 'L6', 'H6', 'S6', 'V6', 'X6', 'D6', 'N6',
    'DH1', 'DH2', 'DH3'
]

df
```

```python
print(" ĐANG TIẾN HÀNH ĐIỀN GIÁ TRỊ MEAN CHO CÁC Ô TRỐNG... ")

# Vòng lặp duyệt qua từng cột điểm số để tính mean và điền khuyết trực tiếp
for cot in cac_cot_diem:
    if cot in df.columns:
        df[cot].fillna(df[cot].mean(), inplace=True)

print(" XỬ LÝ HOÀN TẤT! ")
```

```python
print(" SỐ LƯỢNG DỮ LIỆU THIẾU SAU KHI XỬ LÝ ")
df
```

```python
print(" TẠO BIẾN ĐIỂM TRUNG BÌNH MÔN CÁC NĂM ")
df['TBM1'] = (df['T1']*2 + df['L1'] + df['H1'] + df['S1'] + df['V1']*2 + df['X1'] + df['D1'] + df['N1']) / 10
df['TBM2'] = (df['T2']*2 + df['L2'] + df['H2'] + df['S2'] + df['V2']*2 + df['X2'] + df['D2'] + df['N2']) / 10
df['TBM3'] = (df['T6']*2 + df['L6'] + df['H6'] + df['S6'] + df['V6']*2 + df['X6'] + df['D6'] + df['N6']) / 10
df
```

```python
print(" XẾP LOẠI HỌC LỰC CỦA HỌC SINH ")
def xep_loai(tbm):
    if tbm < 5.0: return 'Y'
    elif tbm < 6.5: return 'TB'
    elif tbm < 8.0: return 'K'
    elif tbm < 9.0: return 'G'
    else: return 'XS'

df['XL1'] = df['TBM1'].apply(xep_loai)
df['XL2'] = df['TBM2'].apply(xep_loai)
df['XL3'] = df['TBM3'].apply(xep_loai)
df
```

```python
print(" QUY ĐỔI SANG THANG ĐIỂM 4 CỦA MỸ ")
df['US_TBM1'] = 4 * (df['TBM1'] / 10)
df['US_TBM2'] = 4 * (df['TBM2'] / 10)
df['US_TBM3'] = 4 * (df['TBM3'] / 10)
df
```

```python
print("KẾT QUẢ TUYỂN SINH")
def xet_tuyen(row):
    khoi = str(row['KT']).strip().upper()
    dh1, dh2, dh3 = row['DH1'], row['DH2'], row['DH3']

    if khoi in ['A', 'A1']: diem_xt = (dh1*2 + dh2 + dh3) / 4
    elif khoi == 'B': diem_xt = (dh1 + dh2*2 + dh3) / 4
    else: diem_xt = (dh1 + dh2 + dh3) / 3

    return 1 if diem_xt >= 5.0 else 0

df['KQXT'] = df.apply(xet_tuyen, axis=1)
df
```
