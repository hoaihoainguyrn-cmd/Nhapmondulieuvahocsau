```python
import pandas as pd
import numpy as np
df = pd.read_csv('processed_dulieuxettuyendaihoc.csv')
```

```python
df_sort2 = df.sort_values(by=['GT', 'DH2'], ascending=[True, True])
df_sort2
```

```python
def q1(x):
    return x.quantile(0.25)

def q3(x):
    return x.quantile(0.75)

pivot_3 = df.groupby('KT').agg(
    count=('DH1', 'count'),
    sum=('DH1', 'sum'),
    mean=('DH1', 'mean'),
    median=('DH1', 'median'),
    min=('DH1', 'min'),
    max=('DH1', 'max'),
    std=('DH1', 'std'),
    Q1=('DH1', q1),
    Q2_median=('DH1', 'median'),
    Q3=('DH1', q3)
)

display(pivot_3)
```

```python
pivot_4 = df.groupby(['KT', 'KV']).agg(
    count=('DH1', 'count'),
    sum=('DH1', 'sum'),
    mean=('DH1', 'mean'),
    median=('DH1', 'median'),
    min=('DH1', 'min'),
    max=('DH1', 'max'),
    std=('DH1', 'std'),
    Q1=('DH1', q1),
    Q2_median=('DH1', 'median'),
    Q3=('DH1', q3))
display(pivot_4)
```

```python
df_theo_KT_KV_DT = df.groupby(['KT', 'KV', 'DT']).agg(
    count=('DH1', 'count'),
    sum=('DH1', 'sum'),
    mean=('DH1', 'mean'),
    median=('DH1', 'median'),
    min=('DH1', 'min'),
    max=('DH1', 'max'),
    std=('DH1', 'std'),
    Q1=('DH1', q1),
    Q2_median=('DH1', 'median'),
    Q3=('DH1', q3)
)
display(df_theo_KT_KV_DT)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tan_so = df['GT'].value_counts()
tan_suat = df['GT'].value_counts(normalize=True) * 100
bang_gt = pd.DataFrame({'Tần số': tan_so, 'Tần suất (%)': tan_suat})
print(bang_gt)


fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.countplot(x='GT', data=df, ax=axes[0], palette='Set2')
axes[0].set_title('Biểu đồ tần số (Cột) - Giới tính')
axes[0].set_xlabel('Giới tính')
axes[0].set_ylabel('Số lượng')
axes[1].pie(tan_suat, labels=tan_suat.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
axes[1].set_title('Biểu đồ tần suất (Tròn) - Giới tính')
plt.tight_layout()
plt.show()
```

```python
cac_bien_tbm = ['US_TBM1', 'US_TBM2', 'US_TBM3']
print(df[cac_bien_tbm].describe())
plt.figure(figsize=(15, 4))
for i, bien in enumerate(cac_bien_tbm, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[bien], kde=True, color='skyblue')
    plt.title(f'Phân phối điểm {bien}')
    plt.xlabel('Điểm')
    plt.ylabel('Số lượng')

plt.tight_layout()
plt.show()
```

```python
bins = [0, 5, 7, 8, float('inf')]
labels = ['k', 'tb', 'kh', 'g']
df['phanlopt1'] = pd.cut(df['T1'], bins=bins, labels=labels, right=False)

df_pivot = df.pivot(columns='phanlopt1', values='T1')

plt.figure(figsize=(15, 6))
colors = ['#e74c3c', '#f39c12', '#3498db', '#2ecc71']
labels_dict = {'k': 'Nhóm K', 'tb': 'Nhóm TB', 'kh': 'Nhóm KH', 'g': 'Nhóm G'}

for i, col in enumerate(labels):
    if col in df_pivot.columns:
        plt.plot(df_pivot.index, df_pivot[col], marker='o', linestyle='-',
                 linewidth=1.5, markersize=5, color=colors[i], label=labels_dict[col])

plt.title('Biểu đồ đường Multiple Line dựa trên số thứ tự học sinh gốc', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Thứ tự học sinh gốc (Index)', fontsize=12)
plt.ylabel('Điểm môn Toán (T1)', fontsize=12)

plt.grid(axis='both', linestyle='--', alpha=0.5)
plt.legend(title="Phân lớp học lực", loc='upper right')

plt.tight_layout()
plt.show()
```

```python
bins = [0, 5, 7, 8, float('inf')]
labels = ['k', 'tb', 'kh', 'g']
df['phanlopt1'] = pd.cut(df['T1'], bins=bins, labels=labels, right=False)

colors = {'k': '#e74c3c', 'tb': '#f39c12', 'kh': '#3498db', 'g': '#2ecc71'}
labels_dict = {'k': 'Nhóm Kém', 'tb': 'Nhóm Trung Bình', 'kh': 'Nhóm Khá', 'g': 'Nhóm Giỏi'}

plt.figure(figsize=(16, 7))

for nhom in labels:
    df_nhom = df[df['phanlopt1'] == nhom]

    plt.vlines(x=df_nhom.index, ymin=0, ymax=df_nhom['T1'],
               colors=colors[nhom], linestyle=':', linewidth=1, alpha=0.6)

    plt.scatter(df_nhom.index, df_nhom['T1'],
                color=colors[nhom], label=labels_dict[nhom], s=40, zorder=3)

plt.title('Biểu đồ Drop-line phân phối điểm T1 theo nhóm học lực', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Thứ tự học sinh gốc (Index)', fontsize=12)
plt.ylabel('Điểm môn Toán (T1)', fontsize=12)
plt.ylim(0, 10.5)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.legend(title="Phân lớp học lực", loc='upper right')

plt.tight_layout()
plt.show()
```
